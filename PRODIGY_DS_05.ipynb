```python
#storing categorical column names to a new variable
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
#for categorical values we can replace the null values with the Mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

**Identifying pattern related to Road conditions, Weather and Time of the day**

```python
# Count accidents by road conditions
road_condition_counts = df_subset['Road_surface_conditions'].value_counts()

# Display the counts
print(road_condition_counts)
```

```python
# Plot road condition counts
plt.figure(figsize=(8, 6))
sns.barplot(x=road_condition_counts.values, y=road_condition_counts.index, palette='Blues')
plt.ylabel('Road Surface Conditions')
plt.xlabel('Number of Accidents')
plt.title('Accidents by Road Surface Conditions')
plt.show()
```

```python
# Count accidents by weather conditions
weather_condition_counts = df_subset['Weather_conditions'].value_counts()

# Display the counts
print(weather_condition_counts)
```

```python
# Plot weather condition counts
plt.figure(figsize=(12, 6))
sns.barplot(x=weather_condition_counts.values, y=weather_condition_counts.index, palette='Blues')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Conditions')
plt.title('Accidents by Weather Conditions')
plt.show()
```

```python
# Extract hour from 'Time'
df['Hour'] = pd.to_datetime(df['Time'], errors='coerce').dt.hour
df.dropna(subset=['Hour'], inplace=True)  # Drop rows with invalid times
```

```python
# Accidents by hour of the day
plt.figure(figsize=(10, 6))
sns.histplot(df['Hour'], bins=24, kde=True, palette='Blues')
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()
```

**Visualizing Accident spots and Contributing factors**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Road_surface_conditions_encoded'] = le.fit_transform(df['Road_surface_conditions'])
df['Weather_conditions_encoded'] = le.fit_transform(df['Weather_conditions'])
df['Time_encoded'] = le.fit_transform(df['Time'])
```

```python
# Visualizing accident spots based on 'Area_accident_occured'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Area_accident_occured', palette='Blues')
plt.title('Number of Accidents by Area', fontsize=14)
plt.xlabel('Area Where Accident Occurred', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Visualizing accident spots based on 'Area_accident_occured'
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Area_accident_occured', hue='Accident_severity', palette='Blues')
plt.title('Accidents by Area and Severity', fontsize=14)
plt.xlabel('Area Where Accident Occurred', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Accident Severity', title_fontsize='10', fontsize='10')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Correlation matrix for the encoded factors
encoded_factors = df[['Road_surface_conditions_encoded', 'Weather_conditions_encoded', 'Time_encoded']]
encoded_factors.columns = ['Road Surface Conditions', 'Weather Conditions', 'Time']
correlation_matrix = encoded_factors.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='Blues')
plt.title('Correlation Matrix of Contributing Factors')
plt.show()
```
